```python
import numpy as np
import random
from skimage import io
from skimage.draw import line
from skimage.morphology import square, closing, dilation
from skimage.transform import rotate, rescale
from skimage.util import img_as_bool, img_as_float, pad, invert
%matplotlib inline

def random_img_pair():
    start = (150,0)

    size = (300,300)
    points = []
    x = start[1]
    y = start[0]
    points.append((y,x))

    while x < size[1]:
        next_x = random.randrange(40) + x
        while True:
            next_y = random.choice([-1,1]) * random.randrange(20,40) + y
            if next_y < size[0] and next_y > 0:
                break
                
        points.append((next_y,next_x))
        x = next_x
        y = next_y
        
    last_point = points[-1]
    points[-1] = (last_point[0], 299)
    mask = np.zeros(size)
    for i in range(0,len(points)-1):
        rr, cc = line(points[i][0], points[i][1], points[i+1][0], points[i+1][1])
        mask[rr, cc] = 1.0
    mask = np.rot90(mask)
    mask = dilation(mask,square(3))
    for col in mask:
        a_max = np.argmax(col)
        col[0:a_max+1] = 1.0
    return mask, invert(mask)

def random_transformations(img, angle):
    result = rotate(img,angle, resize=True)
    rescale_factor = random.random()
    result = rescale(result, 0.5 if rescale_factor < 0.5 else rescale_factor , mode = 'constant')
    result = pad(result, (20,20), 'constant', constant_values=(0, 0))
    return result

img_left, img_right = random_img_pair()
img_left = random_transformations(img_left,10)
img_right = random_transformations(img_right, 10)

io.imshow(img_left)
```

```python
path = "data/set9/"
file = open(path + "correct.txt", mode = "w")
size = 10000
for i in range(0,size):
    if i % 10 == 0:
        print("{}/{}".format(i,size))
    img_left, img_right = random_img_pair()
    img_left = random_transformations(img_left, random.randrange(0,360))
    io.imsave(path + str(i) + ".png", img_left)
    for j in range(1, 73):
        img_right_cp = random_transformations(img_right, j*5)
        io.imsave(path + str(i) + str(j) + ".png",img_right_cp)
        file.write(str(i) + str(j) + " ")
    file.write("\n")
file.close()
```
